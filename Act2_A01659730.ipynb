```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/content/serious-injury-outcome-indicators-2000-2023.csv')

sub = df[['Lower_CI', 'Upper_CI']].dropna()

corr = sub.corr()

plt.figure(figsize=(4, 3))
sns.heatmap(
    corr,
    cmap='coolwarm',
    linewidths=0.5
)
plt.title('Correlación: Lower_CI vs Upper_CI')
plt.tight_layout()
plt.show()
# estan casi perfectamente correlacionadas
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the number of points per cluster
num_points = 100

# Define the centers of the clusters
cluster_centers = np.array([[2, 2], [8, 2], [2, 8], [8, 8]])

# Generate random points for each cluster
data_points = []
for center in cluster_centers:
    points = np.random.randn(num_points, 2) * 1.5 + center  # Adjust 1.5 for cluster spread
    data_points.extend(points)

# Combine the points into a single array
data_points = np.array(data_points)

# Plot the point cloud
plt.scatter(data_points[:, 0], data_points[:, 1], s=10)
plt.title("2D Point Cloud with 4 Clusters")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

1 Initialize
	Choose the number of clusters (K) and randomly select K data points as initial centroids

2 Assignment
	Assign each data point to the nearest centroid based on Euclidean distance

3 Update
	Recalculate the centroids by finding the mean of all data points assigned to each cluster

4 Iteration
	Repeat steps 2 and 3 until the cluster assignment no longer changes significantly (you set the threshold!)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min
from mpl_toolkits.mplot3d import Axes3D

plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')

dataframe = pd.read_csv('/content/analisis.csv')

print(dataframe.describe())
print(dataframe.groupby('categoria').size())

dataframe.drop(['categoria'], axis=1).hist()
sb.pairplot(dataframe.dropna(), hue='categoria', height=4, vars=["op", "ex", "ag"], kind='scatter')
plt.show()
```

```python
X = np.array(dataframe[["op", "ex", "ag"]])
y = np.array(dataframe['categoria'])
#y = np.array(dataframe['year'])
X.shape
y

fig = plt.figure()
ax = Axes3D(fig)
colores=['blue','red','green','blue','cyan','yellow','orange','black','pink','brown','purple']
asignar = []
for row in y:
    asignar.append(colores[row])
    #asignar.append(colores[row-2015])
ax.scatter(X[:,0], X[:,1], X[:,2], c=asignar, s=60)
```

```python
#Valor de K
Nc = range(1, 20)
kmeans = [KMeans(n_clusters=i) for i in Nc]
kmeans
score = [kmeans[i].fit(X).score(X) for i in range(len(kmeans))]
score
plt.plot(Nc,score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()

#Using K=5
kmeans = KMeans(n_clusters=5).fit(X)
centroids = kmeans.cluster_centers_
print(centroids)

labels = kmeans.predict(X)
C = kmeans.cluster_centers_
colores=['red','green','blue','cyan','yellow']
asignar=[]
for row in labels:
    asignar.append(colores[row])
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:,0], X[:,1], X[:,2], c=asignar, s=60)
ax.scatter(C[:,0], C[:,1], C[:,2], marker='*', c=colores, s=1000)
```

```python
f1 = dataframe['op'].values
f2 = dataframe['ex'].values

plt.scatter(f1, f2, c=asignar, s=70)
plt.scatter(C[:,0], C[:,1], marker="*", c=colores, s=1000)
plt.show()
```
